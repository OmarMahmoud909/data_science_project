```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("desktop/Anime.csv")
```

```python
data
```

# Data Cleaning

```python
#shows how many cells are empty in each column
data.isnull().sum()
```

```python
# drop unsed(unimportant) Columns 
data.drop(['End_year','Related_anime','staff'], axis=1, inplace=True)
```

```python
#print Columns after removing the three columns
print(data.columns.tolist())
```

```python
# Number of rows (now)
print("Current Number of rows", data.shape[0])
# check Rows Duplication
duplicated_rows = sum(data.duplicated())
print("Number of Duplicated rows: ", duplicated_rows)
if(duplicated_rows):
    data.drop_duplicates(keep ='first', inplace=True)
print("New Number of rows after Removing duplication", data.shape[0])
```

```python
#delete unimportant data in "type" column
data.drop(data.index[data['Type'] == 'Music'], inplace = True)
```

```python
data.head(88)
```

```python
# Fill missing values in each column with a specific value
data['Japanese_name'].fillna('unknown', inplace=True)
data['Studio'].fillna('unknown', inplace=True)
data['Release_season'].fillna('unknown', inplace=True)
data['Content_Warning'].fillna('unknown', inplace=True)
data['Description'].fillna('unknown', inplace=True)
data['Tags'].fillna('unknown', inplace=True)
data['Related_Mange'].fillna('unknown', inplace=True)
data['Voice_actors'].fillna('unknown', inplace=True)
```

```python
# Remove rows with missing values
data.dropna(subset=['Rating'], inplace=True)
data.dropna(subset=['Episodes'], inplace=True)
data.dropna(subset=['Release_year'], inplace=True)
```

# Q2:Information about the Distribution of Rating and Release_year

```python
#Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['Rating'], bins=30, kde=True)
plt.title('Distribution of Rating')
plt.show()
plt.figure(figsize=(10, 6))
sns.histplot(data['Release_year'], bins=30, kde=True)
plt.title('Distribution of Release_year')
plt.show()
```

# Q3: Relationship between Release_year and Rating

```python
#using Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Release_year', y='Rating', data=data)
plt.title('Scatter Plot between Release_year and Rating')
plt.show()
```

# Q4: Information about the Rating using "boxplot"

```python
sns.boxplot(x='Rating', data=data)
plt.show()
```

# Q5: Information about the avrage of Ranig and Release_year

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2);
# Plot violin plot on axes 1
ax1.violinplot(data.Rating, showmedians=True)
ax1.set_title('Rating')
ax1.set_xticks([1])

# Plot violin plot on axes 2
ax2.violinplot(data.Release_year, showmedians=True)
ax2.set_title('Release_year')
ax2.set_xticks([1])
plt.show()
```

# Q6: Information about types of Anime 

```python
# Count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Type', data=data)
plt.title('Count Plot for type')
plt.show()
```

# Q7: Correlation matrix between Rating ,Release_year, Episodes


```python
# construct data_ from three columns of the data (Rating ,Release_year, Episodes)
data_ = data[['Rating','Release_year', 'Episodes']]
correlation_matrix = data_.corr()
#heatmap
sns.heatmap(correlation_matrix, cmap='viridis' , annot = True)
#set title
plt.title("Correlation matrix for Rating ,Release_year, Episodes")
# show plot
plt.show()
```
